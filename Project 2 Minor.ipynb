```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(scaled_data)
similarity_matrix
```

```python
scaled_data
```

```python
def recommend(song_index, top_n=5):
    sim_scores = list(enumerate(similarity_matrix[song_index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:top_n+1]  # exclude itself
    song_indices = [i[0] for i in sim_scores]
    return df.iloc[song_indices]
```

```python
print("Recommendations for first song in dataset:")
print(recommend(0))
```
